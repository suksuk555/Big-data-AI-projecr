```python
import pandas as pd
```

```python
df = pd.read_csv('data\pet_dataset_final.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df["Budget"] = pd.cut(
    df["Budget"],
    bins=[-1, 1000, 3000, float("inf")],    
    labels=["Less than 1000", "Between 1000 and 3000", "Greater than 3000"]
)

print(df.head())

```

```python
df.head()
```

```python
df["PreferQuietPet"] = df["PreferQuietPet"].replace(1, "Yes")
df["PreferQuietPet"] = df["PreferQuietPet"].replace(0, "No")


df["Allergies"] = df["Allergies"].replace(1, "Yes")
df["Allergies"] = df["Allergies"].replace(0, "No")
```

```python
df
```

```python
df.to_csv("data_ready/dataset1.csv", index=False, encoding="utf-8-sig")
```
