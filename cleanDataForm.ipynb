```python
df.rename(columns={"สัตว์เลี้ยงที่เลี้ยงอยู่คืออะไร": "PetChoice",
                   "เพศ":"Gender",
                   "งบประมาณในการเลี้ยงสัตว์(ต่อเดือน)":"Budget",
                   "เวลาว่างเฉลี่ยต่อวัน (24 ชั่วโมง)":"FreeTime",
                   "เวลาอยู่บ้านเฉลี่ยต่อวัน (24 ชั่วโมง)":"TimeAtHome",
                   "ที่อยู่อาศัย": "Living",
                   "สัตว์เลี้ยงเงียบหรือไม่":"PreferQuietPet",
                   "เป็นภูมแพ้หรือไม่":"Allergies",


                   
                   
                   
                    },inplace=True)
```

```python
# แยกคอลัมน์ PetChoice ออกมา
pet_choice = df.pop('PetChoice')

# แทรกกลับเข้าไปหลัง Allergies
df.insert(df.columns.get_loc('Allergies') + 1, 'PetChoice', pet_choice)
```

Column Gender

```python
df['Gender'] = df['Gender'].replace({'ชาย': 'Male', 'หญิง': 'Female'})
```

```python
df
```

Column PetChoice

```python
df['PetChoice'] = df['PetChoice'].replace({
    'ปลา': 'Fish',
    'สุนัข': 'Dog',
    'แมว': 'Cat',
    'แฮมเตอร์': 'Hamster',
    'งู': 'Snake',
    'ไม่เลี้ยงสัตว์': 'No pet'
})
```

```python
df
```

Column Freetime,TimeAtHome

```python
def extract_numeric(val):
    try:
        return int(''.join(filter(str.isdigit, str(val))))
    except:
        return None

df['FreeTime'] = df['FreeTime'].apply(extract_numeric)
df['TimeAtHome'] = df['TimeAtHome'].apply(extract_numeric)

ชdf = df[(df['FreeTime'] <= 24) & (df['TimeAtHome'] <= 24)]
```

```python
df
```

Column Budget

```python
# ฟังก์ชันแปลงงบประมาณเป็นตัวเลข
def extract_budget(val):
    try:
        return int(''.join(filter(str.isdigit, str(val))))
    except:
        return None

# สร้างคอลัมน์ชั่วคราวเพื่อใช้กรอง
df['__budget_numeric'] = df['Budget'].apply(extract_budget)

# ลบแถวที่งบประมาณเกิน 100000
df = df[df['__budget_numeric'] <= 100000]

# แทนค่าในคอลัมน์ Budget เดิมด้วยช่วงที่ต้องการ
def categorize_budget(num):
    if num is None:
        return 'Unknown'
    elif num < 1000:
        return 'Less than 1000'
    elif 1000 <= num <= 3000:
        return 'Between 1000 and 3000'
    else:
        return 'Greater than 3000'

df['Budget'] = df['__budget_numeric'].apply(categorize_budget)

# ลบคอลัมน์ชั่วคราว
df.drop(columns=['__budget_numeric'], inplace=True)
```
